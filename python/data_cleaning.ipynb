```python
import pandas as pd

df = pd.read_excel("../data/raw/Financial Sample.xlsx")

df.columns = df.columns.str.strip()

df["Discount Band"] = df["Discount Band"].fillna("Unknown")
```

```python
country_sales = df.groupby("Country")["Sales"].sum().sort_values(ascending=False)

country_sales
```

```python
import matplotlib.pyplot as plt
```

```python
country_sales.plot(kind="bar", figsize=(8,5))

plt.title("Sales by Country")
plt.xlabel("Country")
plt.ylabel("Sales")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(8,5))
country_sales.plot(kind="bar")

plt.title("Sales by Country")
plt.xlabel("Country")
plt.ylabel("Sales")

plt.tight_layout()

plt.savefig("../dashboard_images/sales_by_country.png")

plt.show()
```

```python
monthly_sales = df.groupby("Month Name")["Sales"].sum()

plt.figure(figsize=(10,5))
monthly_sales.plot(kind="line", marker="o")

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.tight_layout()

plt.savefig("../dashboard_images/monthly_sales.png")

plt.show()
```
