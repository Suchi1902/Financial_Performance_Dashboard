# Business Insights

## Objective
Analyze the financial performance of the company using sales, profit, and customer segment data.

## Questions
- Which country has the highest sales?
- Which product generates the highest profit?
- What is the monthly sales trend?
- Which customer segment contributes the most sales?
- What is the overall profit margin?

## Key Performance Indicators (KPIs)

In this section, we calculate the main business metrics:

- Total Sales
- Total Profit
- Total Units Sold
- Average Discount

```python
# Key Performance Indicators

total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
total_units = df["Units Sold"].sum()
avg_discount = df["Discounts"].mean()

print(f"Total Sales      : {total_sales:,.2f}")
print(f"Total Profit     : {total_profit:,.2f}")
print(f"Units Sold       : {total_units:,.0f}")
print(f"Average Discount : {avg_discount:,.2f}")
```

## Country-wise Sales Analysis

This analysis identifies which countries generate the highest revenue.

```python
country_sales = (
    df.groupby("Country")["Sales"]
      .sum()
      .sort_values(ascending=False)
)

country_sales
```

### Interpretation

- The analysis shows the sales performance across different countries.
- This helps identify the strongest markets.
- The company can focus marketing efforts on high-performing countries while improving sales strategies in lower-performing regions.

## Product-wise Profit Analysis

This analysis identifies the products that generate the highest profit.

```python
product_profit = (
    df.groupby("Product")["Profit"]
      .sum()
      .sort_values(ascending=False)
)

product_profit
```

```python
plt.figure(figsize=(10,5))

product_profit.plot(kind="bar")

plt.title("Profit by Product")
plt.xlabel("Product")
plt.ylabel("Profit")

plt.tight_layout()

plt.savefig("../dashboard_images/profit_by_product.png")

plt.show()
```

### Interpretation

- This chart shows which products generate the highest profit.
- High-profit products should receive greater marketing and inventory focus.
- Low-profit products may require pricing or cost optimization.

## Customer Segment Analysis

This analysis compares sales across different customer segments.

```python
segment_sales = df.groupby("Segment")["Sales"].sum()

segment_sales
```

```python
plt.figure(figsize=(6,6))

segment_sales.plot(kind="pie", autopct="%1.1f%%")

plt.ylabel("")

plt.title("Sales by Segment")

plt.savefig("../dashboard_images/sales_by_segment.png")

plt.show()
```

## Monthly Sales Trend

This analysis examines how sales change over different months to identify seasonal trends and business patterns.

```python
# Arrange months in calendar order
month_order = [
    "January", "February", "March", "April", "May", "June",
    "July", "August", "September", "October", "November", "December"
]

monthly_sales = (
    df.groupby("Month Name")["Sales"]
      .sum()
      .reindex(month_order)
)

monthly_sales
```

```python
plt.figure(figsize=(10,5))

monthly_sales.plot(marker="o", linewidth=2)

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.grid(True)

plt.tight_layout()

plt.savefig("../dashboard_images/monthly_sales.png")

plt.show()
```

### Interpretation

- This chart highlights monthly sales performance.
- Peak months indicate periods of high customer demand.
- Lower-performing months may require promotional campaigns.
- Understanding seasonality helps improve forecasting and inventory planning.
